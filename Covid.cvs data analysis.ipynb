```python
def iqr_outlier(data):
    for i in range(len(data.iloc[1])):
        if type(data.iloc[0, i]) != np.int64 and type(data.iloc[0, i]) != np.float64:
            continue
        q1, q3 = np.percentile(data.iloc[:, i], [25, 75])
        print(f"q1: {q1}")
        print(f"q3: {q3}")

        iqr = q3 - q1
        lower = q1 - (iqr * 1.5)
        upper = q3 + (iqr * 1.5)
        print(f"lower: {lower}")
        print(f"upper: {upper}")

        data = data[data.iloc[:, i] < upper]
        data = data[data.iloc[:, i] > lower]
        
    return data

covid_data = iqr_outlier(covid_data)
```

```python
plt.scatter(covid_data['total_vaccinations'],covid_data['hosp_patients'])
plt.xlabel('total vaccinations')
plt.ylabel('hosp patients')
plt.show()
```

```python
X = covid_data.total_vaccinations.values
Y = covid_data.hosp_patients.values

cov = np.cov([X, Y])[0][1]
ccef = np.corrcoef([X, Y])[0][1]

print(f"공분산: {cov}, 상관 계수: {ccef}")
```

```python
#결과해석:
#0 < 𝑟 < 0.4 – 희박한 양적 상관 관계 ~ 낮은 양적 상관 관계
#0.4 < 𝑟 < 0.8 – 비교적 높은 양적 상관 관계 ~ 높은 양적 상관 관계
#0.8 < 𝑟 < 1 – 매우 높은 양적 상관 관계

# => 상관 계수가 0.56이므로 높은 양적 상관 관계가 있다.
```

```python
#귀무 가설: 백신을 맞은 사람의 수와 병원 환자 수 사이에 상관 관계가 없다
#대립 가설: 백신을 맞은 사람의 수와 병원 환자 수 사이에 상관 관계가 있다
```
