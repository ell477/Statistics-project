## 단순 선형 회귀 분석

- 회귀 분석에서, 설명하고자 하는 변수는 종속 변수이다.
- 회귀 분석에서, 설명하기 위해 사용하는 변수는 독립 변수이다.
- 단순 선형 회귀 분석에서는 **가장 적합한** 모델을 찾기 위해 최소 제곱법을 사용한다.
- 결정 계수가 높을수록 모델이 현상을 잘 설명하는 것이다.

### 단순 선형 회귀 분석 실습해보기
다음은 당뇨병 환자에 관한 데이터의 일부이다. BMI를 이용해 당뇨 수치를 설명하는 모델을 제안하고, 적절히 검정하시오.

```python
dia = datasets.load_diabetes()

dia = pd.DataFrame(dia['data'], index=dia['target'], columns=dia['feature_names'])
display(dia)
```

```python
dia = dia['bmi']
display(dia)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(dia.values.reshape(-1, 1), dia.index.to_numpy().reshape(-1, 1))

print(f"모델: y = {lr.intercept_} + {lr.coef_[0]}X + e")
```

```python
#귀무 가설: 독립 변수와 종속 변수 사이에는 선형적 관계가 없다.
#대립 가설: 독립 변수와 종속 변수 사이에는 선형적 관계가 있다.
```

```python
import statsmodels.api as sm

results = sm.OLS(dia.index, sm.add_constant(dia)).fit()
results.summary()
```

```python
#해석
#bmi에 관한 p-value가 0.000이므로 100% 신뢰 수준에서 귀무 가설을 기각할 근거가 충분하다.
# => 두 변수 사이에는 선형적 관계가 존재한다!
```

## 다중 선형 회귀 분석

- 회귀 분석에서, 설명하고자 하는 변수는 종속 변수이다.
- 회귀 분석에서, 설명하기 위해 사용하는 변수는 독립 변수이다.
- 조정된 결정 계수는 일반 결정 계수보다 항상 낮다.
- F 검정은 모델을 전체적으로 평가한다.

### 다중 선형 회귀 분석 실습해보기
다음은 당뇨병 환자에 관한 데이터의 일부이다. 나이, BMI, 혈압(bp)을 모두 이용해 당뇨 수치를 설명하는 모델을 제안하고, 적절히 검정하시오.

```python
dia = datasets.load_diabetes()

dia = pd.DataFrame(dia['data'], index=dia['target'], columns=dia['feature_names'])
dia = dia.reset_index().rename(columns={'index': 'dia'})
display(dia)
```

```python
dia = dia[['age','bmi','bp']]
display(dia)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(dia.values, dia.index.to_numpy().reshape(-1, 1))

print(f"모델: y = {lr.intercept_[0]} + {lr.coef_[0][0]} (age) + {lr.coef_[0][0]} (BMI) + {lr.coef_[0][1]} (bp) + e")
```

```python
#귀무 가설: 독립 변수와 종속 변수 사이에는 선형적 관계가 없다.
#대립 가설: 독립 변수와 종속 변수 사이에는 선형적 관계가 있다.
```

```python
import statsmodels.api as sm

results = sm.OLS(dia.index, sm.add_constant(dia)).fit()
results.summary()
```

```python
#해석
#나이(age): p-value가 0.113이므로 귀무 가설을 기각할 근거가 충분하지 않다.
# => 나이와 당뇨 수치 사이에는 선형적 관계가 없다.
#BMI: p-value가 0.686이므로 귀무 가설을 기각할 근거가 충분하지 않다.
# => BMI와 당뇨 수치 사이에는 선형적 관계가 없다.
#혈압(bp): p-value가 0.379이므로 귀무 가설을 기각할 근거가 충분하지 않다.
# => 혈압과 당뇨 수치 사이에는 선형적 관계가 없다.
```
